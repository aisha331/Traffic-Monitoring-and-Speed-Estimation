# Importing libraries


```python
import cv2
import time 
import numpy as np
```

# Initializing parameters 

```python
width_min=80 
height_min=80 
offset=6 

x1_line1=25
x2_line1=1150
y_line1=550 
x1_line2=450
x2_line2=820
y_line2=350

delay= 60 
detect = []
count= 0
start_time =time.time()
distance= 30    
```

# Speed function

```python
def Speed_Cal(time):
    try:
        Speed=(distance/1000)/(time/3600)
        return int(Speed)
    except ZeroDivisionError:
        print("can not devide by Zero")
```

# Car center function 

```python
def pos_center_car(x, y, w, h):
    x1 = int(w / 2)
    y1 = int(h / 2)
    cx = x + x1
    cy = y + y1
    return cx,cy   
```

# Importing video and Object detection

```python
cap = cv2.VideoCapture('C:\\Users\\97150\\Desktop\\Y3-TERM2\\year2-term2\\cv project-alex]\\video.mp4')
subtracting = cv2.bgsegm.createBackgroundSubtractorMOG()
```

# Cars counting and Speed estimation

```python
while True:
    ret , frame1 = cap.read()
    timee = float(1/delay)
    time.sleep(timee) 
    
    grey = cv2.cvtColor(frame1,cv2.COLOR_RGB2GRAY)
    blur = cv2.GaussianBlur(grey,(3,3),5)
    img_sub = subtracting.apply(blur)
    dilat = cv2.dilate(img_sub,np.ones((5,5))) 
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))  
    dilation = cv2.morphologyEx (dilat, cv2. MORPH_CLOSE , kernel) 
    dilation = cv2.morphologyEx (dilation, cv2. MORPH_CLOSE , kernel)
    Contour,h=cv2.findContours(dilation,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE) 
    

    cv2.line(frame1, (x1_line2, y_line2), (x2_line2, y_line2), (255,127,0), 3) 
    cv2.line(frame1, (x1_line1, y_line1), (x2_line1, y_line1), (255,127,0), 3)
    
    for(i,c) in enumerate(Contour):
        (x,y,w,h) = cv2.boundingRect(c)
        valid_Contour = (w >= width_min) and (h >= height_min)
        if not valid_Contour:
            continue

        cv2.rectangle(frame1,(x,y),(x+w,y+h),(0,255,0),2) 
        center = pos_center_car(x, y, w, h)
        detect.append(center)
        cv2.circle(frame1, center, 4, (0, 0,255), -1)


        for (x,y) in detect:
            if y<(y_line1+offset) and y>(y_line1-offset):
                count+=1
                cv2.line(frame1, (x1_line1, y_line1), (x2_line1, y_line1), (255,0,0), 5)  
                detect.remove((x,y))
                print("car is detected : "+str(count))        
       
        while int(y_line2) >= int(x+(h/2)):
            cv2.line(frame1, (x1_line2, y_line2), (x2_line2, y_line2), (0,255,0), 3)
            start_time =time.time()
            #print("start_time " + str(int(start_time))+" second ")
            break
        
        while int(y_line2) <= int(x+(h/2)):
            #if int(y_line2) <= int(x+(h/2)) & int(700) >= int(x+(h/2)):
                cv2.line(frame1, (x1_line1, y_line1), (x2_line1, y_line1) ,(0,255,0),3)
                Speed= Speed_Cal(int( time.time()- start_time )+1 )
                print("car number "+ str(count) +" Speed: " +str(Speed)+ " KM/H ")
                cv2.putText(frame1,"Speed: "+str(Speed)+" KM/H", (center[0],center[1]), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255),3)
                break
            #else :
                cv2.putText(frame1, "Calcuting", (500,650), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,0,0),3)
                break

            
    cv2.putText(frame1, "NUMBER OF VEHICLE: "+str(count), (0, 50), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,0,0),2)
    cv2.imshow("Video Original" , frame1)
    cv2.imshow("Detectar",dilation)

    if cv2.waitKey(1) == 27:
        break
    
cv2.destroyAllWindows()
cap.release()
```
